### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
loja['Total'] = loja['Preço'] * loja['Quantidade de parcelas']
loja
```

```python
loja2['Total'] = loja2['Preço'] * loja2['Quantidade de parcelas']
loja2
```

```python
loja3['Total'] = loja3['Preço'] * loja3['Quantidade de parcelas']
loja3
```

```python
loja4['Total'] = loja4['Preço'] * loja4['Quantidade de parcelas']
loja4
```

```python
faturamento = loja['Total'].sum()
faturamento2 = loja2['Total'].sum()
faturamento3 = loja3['Total'].sum()
faturamento4 = loja4['Total'].sum()

faturamentos = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento': [faturamento, faturamento2, faturamento3, faturamento4]
})

faturamentos
```

# 2. Vendas por Categoria

```python
vendas_categoria = loja.groupby('Categoria do Produto')['Total'].sum().reset_index()
vendas_categoria
```

```python
vendas_categoria2 = loja2.groupby('Categoria do Produto')['Total'].sum().reset_index()
print(vendas_categoria2)
```

```python
vendas_categoria3 = loja3.groupby('Categoria do Produto')['Total'].sum().reset_index()
print(vendas_categoria3)
```

```python
vendas_categoria4 = loja4.groupby('Categoria do Produto')['Total'].sum().reset_index()
print(vendas_categoria4)
```

# 3. Média de Avaliação das Lojas


```python
media_aval = round(loja['Avaliação da compra'].mean(), 2)
media_aval2 = round(loja2['Avaliação da compra'].mean(), 2)
media_aval3 = round(loja3['Avaliação da compra'].mean(), 2)
media_aval4 = round(loja4['Avaliação da compra'].mean(), 2)
medias_aval = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Média Avaliação': [media_aval, media_aval2, media_aval3, media_aval4]
})
# print(medias_aval) se quiser s[o visualizar]
medias_aval
```

# 4. Produtos Mais e Menos Vendidos

```python
prod = loja.groupby('Produto')['Produto'].count().sort_values(ascending=False).reset_index(name='Quantidade')
prod_mais = prod.head(1)
prod_menos = prod.tail(1)
prod_resultado = pd.concat([prod_mais, prod_menos])
print(prod_resultado)
```

# 5. Frete Médio por Loja

```python
# iterando
lojas = [loja, loja2, loja3, loja4]
frete_medio = [
    round(l['Frete'].mean(), 2) for l in lojas
]

fretes = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete Médio': frete_medio
})
fretes
```

# 6. Gráficos


```python
import matplotlib.pyplot as plt
```

```python
fretes.set_index("Loja")["Frete Médio"].plot(
    kind="pie",
    autopct="%1.1f%%",   # mostra porcentagens
    figsize=(6,6),       # tamanho do gráfico
    startangle=90        # ângulo inicial
)

plt.title("Distribuição dos Fretes por Cidade")
plt.ylabel("")  # remove o rótulo do eixo y
plt.show()
```

```python
ax = medias_aval.plot(
    kind="bar",
    color="skyblue",
    edgecolor="black",
    figsize=(8,6)
)

# Ajustando limites do eixo Y
ax.set_ylim(3.9, 4.1)

# Adicionando rótulos de dados nas colunas
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", padding=2)

plt.title("Média de Avaliação por Produto")
plt.xlabel("Produtos")
plt.ylabel("Média de Avaliação")
plt.xticks(rotation=45)
plt.show()
```

```python
ax = faturamentos.set_index("Loja")["Faturamento"].plot(
    kind="barh",
    color="lightgreen",
    edgecolor="black",
    figsize=(8,6)
)

def formatar_valor(v):
    return f"{v:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")

# Adicionando rótulos formatados
for container in ax.containers:
    ax.bar_label(container, labels=[formatar_valor(v) for v in faturamentos["Faturamento"]], padding=3)

plt.xlabel("Faturamento (R$)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
